```python
day = 0

q1 = "Is your birthday in set1?\n" + \
    "1  3  5  7\n" + \
    "9  11  13  15\n" +\
    "17  19  21  23\n" + \
    "25  27  29  31" + \
    "\nEnter 0 for No and 1 for Yes: "
answer = eval(input(q1))

if answer == 1:
    day += 1
    
q2 = "Is your birthday in Set2\n?" + \
     "2  3  6  7\n" + \
     "10  11 14  15\n" + \
    "18  19 22  23\n" + \
    "26  27 30  31" + \
    "\nEnter 0 for No and 1 for Yes: "
answer = eval(input(q2))

if answer == 1:
    day +=2

q3 = "Is your bithday in set3?\n" + \
    "4  5  6  7\n" + \
    "12  13  14  15\n" + \
    "20  21  22  23\n"+ \
    "28  29  30  31" + \
    "\nEnter 0 for No and 1 for Yes: "
answer = eval(input(q3))

if answer == 1:
    day += 4
    
q4 = "Is your birthday in set4?\n" + \
     "8  9  10  11\n" + \
     "12  13  14  15\n" + \
     "24  25  26  27\n" + \
    "28  29  30  31" + \
    "\nEnter 0 for No and 1 for Yes: "
answer = eval (input(q4))

if answer == 1:
    day +=8
    
q5 = "Is your birthday in set5?\n" + \
    "16  17  18  19\n" + \
    "20  21  22  23\n" + \
    "24  25  26  27\n" + \
    "28  29  30  31" + \
    "\nEnter 0 for No and 1 for Yes: "
answer = eval(input(q5))

if answer == 1:
    day +=16
    
print("\nYour birthday is " + str(day) + "!")
```
